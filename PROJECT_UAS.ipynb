```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('uas.jpg')
```

```python
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
jeruk_lower = np.array([10, 160, 133], np.uint8)
jeruk_upper = np.array([25, 255, 255], np.uint8)
alpukat_lower = np.array([25, 52, 72], np.uint8)
alpukat_upper = np.array([102, 255, 255], np.uint8)
salak_lower = np.array([6, 50, 5], np.uint8)
salak_upper = np.array([8, 180, 100], np.uint8)
```

```python
jeruk_mask = cv2.inRange(hsv_image, jeruk_lower, jeruk_upper)
alpukat_mask = cv2.inRange(hsv_image, alpukat_lower, alpukat_upper)
salak_mask = cv2.inRange(hsv_image, salak_lower, salak_upper)
```

```python
kernel = np.ones((20,10), np.uint8)
dilated_jeruk_mask = cv2.dilate(jeruk_mask, kernel, iterations=1)
kernel = np.ones((20,10), np.uint8)
dilated_alpukat_mask = cv2.dilate(alpukat_mask, kernel, iterations=1)
kernel = np.ones((20,10), np.uint8)
dilated_salak_mask = cv2.dilate(salak_mask, kernel, iterations=1)
```

```python
jeruk = cv2.bitwise_and(img, img, mask=dilated_jeruk_mask)
alpukat = cv2.bitwise_and(img, img, mask=dilated_alpukat_mask)
salak = cv2.bitwise_and(img, img, mask=dilated_salak_mask)
```

```python
fig, axs = plt.subplots(1,4,figsize=(8,5))


axs[0].imshow(img)
axs[0].set_title("gambar asli")
axs[1].imshow(jeruk)
axs[1].set_title("jeruk")
axs[2].imshow(alpukat)
axs[2].set_title("alpukat")
axs[3].imshow(salak)
axs[3].set_title("salak")
plt.show()
```
